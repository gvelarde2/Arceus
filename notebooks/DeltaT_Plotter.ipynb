```python
#Delta_T Plotter
#VF @RT = 0.72 V
#IF @RT = 0.00050507 A

#K = 0.5598 C/mV typical
#extracts deltaT of diode 

import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt

IF = 0.00050507
K = 559.8
sample = 'D00'
path = 'C:\\Users\\pmlbl\\OneDrive\\Documents\\PythonScripts\\Data\\20241206\\{}'.format(sample)

#Define file directory
dir = os.listdir( path )

headers = ['AHV', 'STS', 'dT']
master = pd.DataFrame(columns = headers)

#Iterate over csv files and populate df matrix
for file in dir:
   if file.endswith(('.csv')):
      #print(file)
      df = pd.read_csv(path+'\\'+file)
      Im = min(enumerate(df.IF), key=lambda x: abs(x[1]-IF))
      Vm = df.VF[Im[0]]
      dV = abs(Vm-0.72)
      AHV = float(file.split("_")[1])
      #print(AHV)
      STS = float(file.split("_")[3])
      dT = dV*K
      sub = pd.DataFrame([AHV,STS,dT]).transpose()
      sub.columns = headers
      master = pd.concat([master,sub])
   else:
      pass

# ###
# import matplotlib.pyplot as plt
# from matplotlib.tri import Triangulation
x = master.AHV.to_numpy()
y = master.STS.to_numpy()
# # fig, ax = plt.subplots()
z = master.dT.to_numpy()

# triang = Triangulation(x,y)
# plt.tricontourf(triang,z,150,cmap='viridis')
# #plt.scatter(x,y)
# plt.colorbar(label='Delta T (C)')
# #plt.tricontour(triang,z,15,colors='k')
# plt.xlabel('Applied Heating Voltage (V)')
# plt.ylabel('Heating Voltage Hold Time (s)')
# plt.show()
# #master
# ##
master
fig, ax = plt.subplots()
#ax.scatter(x,y)
ax.tricontour(x,y,z,150,cmap='viridis',zorder=-1)
t=ax.tricontourf(x,y,z,150,cmap='viridis',zorder=-1)
cb=fig.colorbar(t,label=r'Chiplet $\Delta$Temperature ($^\circ$C)')
ax.set_xlabel('Applied Heating Voltage to Resistor (V)')
ax.set_ylabel('Resistor Heating Voltage Hold Time (s)')
ax.scatter(x,y,s=2,color='k',zorder=1)
ax.set_title('180second Cool Time')
fig.savefig(path)
```

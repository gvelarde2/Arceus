```python
#read all xlsx time vs temp files and plot raw results as well as extracted standard deviation
import pandas as pd 
import os 
import glob 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

class Tstab:
    def __init__(self,path):
        self.path = path

    def get_csvs(self):
        csvs = glob.glob(os.path.join(self.path, "*.csv"))
        return csvs

    def get_df(self):
        df_total = pd.DataFrame()
        for f in self.get_csvs():
            df = pd.read_csv(f,names=['time','temp','setpoint'])
            df.setpoint = int(f.split("\\")[-1].split("C.csv")[0][4:])
            df_total = pd.concat([df_total,df])
        return df_total
    
    def plot_all(self):
        df = self.get_df()
        plt.scatter(df.time,df.temp,c=df.setpoint,cmap='viridis') #plotts only single plot so only one legend can be made
        #print(str(df.setpoint.unique()))
        #st = ['80','90','100','110','120','130']
        #plt.legend(st)
        #plt.legend(df.setpoint.unique())
        plt.colorbar()

# file = []
# STD =[]
# # loop over the list of csv files 
# for f in csv_files: 
    
#     # read the csv file 
#     df = pd.read_csv(f,names=['time','temp','sp']) 
      
#     # print the location and filename 
#     #print('Location:', f)
#     a = file.append(int(f.split("\\")[-1].split("C.csv")[0][4:]))
#     #print('File Name:', f.split("\\")[-1])
#     STD.append(np.std(df.temp[-500:]))
#     plt.plot(df.time,df.temp)
#     plt.xlabel("Time (seconds)")
#     plt.ylabel("Thermocouple Temperature ($^\circ$C)")
#     plt.tick_params(direction='in')
#     plt.grid(visible=True,alpha=0.3)
  
#     #print(f'std is {stat}')
      
#     # print the content 
#     #print('Content:') 
#     # display(df) 
#     # print() 
# print(file)
# plt.legend(file)
# print(STD)
def main():
    path = r'C:\Users\pmlbl\Arceus\data\20241216\D00'
    #df_total = Tstab.get_df(path)
    #display(df_total)
    Tstab(path).plot_all()
if __name__ == "__main__":
    main()
```

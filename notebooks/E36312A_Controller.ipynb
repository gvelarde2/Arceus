```python
import pyvisa
import time
print(pyvisa.ResourceManager().list_resources())
```

```python
PS = pyvisa.ResourceManager().open_resource('USB0::0x2A8D::0x1102::MY61007122::INSTR')
PS.write('SYST:REM\n')

PS.write('APPLY P25V, 20, 0.55\n')
PS.write('OUTPUT:STATE ON\n')
time.sleep(3)
PS.write('OUTPUT:STATE OFF\n')
PS.write('*CLS\n')
PS.write('*RST\n')

```
